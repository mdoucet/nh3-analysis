```python
import numpy as np

from matplotlib import pyplot as plt
import matplotlib.lines as mlines
```

```python
pop = np.loadtxt('population.csv').T
```

```python

yrs = np.arange(1930, 2008)
pop_no_hb = 2000 + (yrs-1930)/50*1000 

yrs_ext = np.arange(2008, 2024)
pop_no_hb_ext = 2000 + (yrs_ext-1930)/50*1000 

fig, ax = plt.subplots(dpi=150, figsize=(5, 4.1))
plt.plot(pop[0], pop[1]/1000000, linewidth=3, label='World population')
plt.plot(yrs, pop_no_hb, linewidth=3, color='green', label='No Haber Bosch')
plt.plot(yrs_ext, pop_no_hb_ext, linewidth=3, linestyle=':', color='green', label='No H-B extrapolated')

plt.legend(loc='lower right', frameon=False, fontsize=10)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population (millions)', fontsize=14)
plt.ylim(0, 8500)
plt.show()


fig, ax = plt.subplots(dpi=150, figsize=(5, 4.1))
wpop = pop[1]/1000000
pop_no_hb = 2000 + (pop[0]-1930)/50*1000 
plt.plot(pop[0], 1-pop_no_hb/wpop, linewidth=3, label='World population')

plt.legend(loc='lower right', frameon=False, fontsize=10)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Population (millions)', fontsize=14)
plt.show()
```
